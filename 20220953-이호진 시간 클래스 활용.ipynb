### 수업 시간표 관리 프로그램 파일 확장 / 2022-06-09 / 20220953-이호진

```python
import copy
import os
import pickle
# 프로그램 제목 출력
print(' [단순한 수업시간표 관리 프로그램]')
print('====================================')

# 수업정보 초기화

class Infor:
    def __init__(self):
        weekdays = ['월', '화', '수', '목', '금'] #요일
        hours    = [ str(x) for x in range(1, 9)] #시간
    def week_bring_num(self):
        print('시간표를 구성할 요일 수 : '+str(weekdays), str(len(weekdays))+"개 입니다")
    def time_bring_num(self):
        print('시간표를 구성할 시간 수 : ', (hours[len(hours)-1:]) ,'교시 입니다 ')
    def quest(self):
        an = int(input('시간표를 구성할 요일 수를 알고싶으면 1, 시간표를 구성할 시간 수를 알고싶으면 2 누르세요 : '))
        if an == 1:
            self.week_bring_num()
        elif an == 2:
            self.time_bring_num()
        else:
            print("잘못누르셨습니다")

classes = {key: { key:' ' for key, value in dict.fromkeys(weekdays).items()} 
                          for key in dict.fromkeys(hours).keys()}         

b = Infor()
b.quest()
```

```python
def print_timetable(**classdict):
    print("\n")
    print("[ TimeTable ]")
    print("-------------")
    print("| |M|T|W|T|F|")                # 요일 출력
    print("-------------")
    
    for hour in classdict.keys():
        print('|', end='')
        
        print(hour, end='|')          # 교시 출력
        for day in classdict[hour].keys():
            print(classdict[hour][day], end='|') # 알파벳 출력
        print()
    print("-------------\n")
```

```python
def dup_check(day, hour, num, **classdict):  # 중복 확인 함수
    dup = False
    for i in range(num):
        if classdict[str(int(hour)+i)][day] != ' ':
            dup = True
            break
    return dup
```

```python
while True:
    
    # 시간표 표시
    print_timetable(**classes)
    
    # 프로그램 메뉴  
    choice = input(">>> 메뉴를 선택하세요 (1. 수업등록, 2. 종료): ")

    if choice == '2':   # 프로그램 종료
        break
    elif choice != '1': # 다른 입력 무시
        print(">>> [ERROR] 잘못 입력하셨습니다!")  # 오류 처리
        continue
    
    load = input(">>> 메뉴를 선택하세요 (1. 파일 불러오기, 2. 불러오지 않기):")
    if load == '1':
        print('1')
        file = 'timetable.pkl'
        if os.path.isfile(file) == True:
            with open('timetable.pkl', 'rb') as file:
                classes = copy.deepcopy(pickle.load(file))
        else:
            print('불러올 시간표가 없습니다.')
            
    # 수업의 요일 및 교시 입력      
    wday, shour = input(">>> 등록할 수업의 요일과 교시를 입력하세요: ").split()
    
    # 요일 입력 오류 체크
    if weekdays.count(wday) == 0:
        print(">>> [ERROR] 요일을 잘못 입력하셨습니다!")      # 요일 입력 오류 
        continue
    
    # 수업시간 범위 체크
        
    if not (1<=int(shour)<=8):
        print(">>> [ERROR] 시작 시간을 잘못 입력하셨습니다!")  # 시작 시간 입력 오류     
        continue
    
    # 수업의 연강 시간 입력  
    count = int(input(">>> 등록할 수업의 연강시간을 입력하세요: "))

    # 연강 시간 범위 체크
    if count < 1:
        print(">>> [ERROR] 연강 시간을 잘못 입력하셨습니다!")  # 연강 시간 입력 오류
        continue
    elif not (1<=int(shour)+count-1<=8):
        print(">>> [ERROR] 등록하려는 수업이 등록 가능 시간을 벗어납니다!")  # 범위 오류
        continue

    # 수업 중복 체크
    if dup_check(wday, shour, count, **classes):
        print(">>> [ERROR] 해당 시간에 이미 수업이 있습니다!")  # 범위 오류    
        continue
        
    # 수업의 알파벳 입력  
    title = input(">>> 등록할 수업의 알파벳을 입력하세요: ")        

    # 수업 정보 입력
    for i in range(count):
        classes[str(int(shour)+i)][wday] = title[0]  # 맨 처음 문자만 입력

save = input(">>> 메뉴를 선택하세요 (1. 텍스트 저장, 2. 딕셔너리 저장 3. 둘 다 저장)")
if save == '1':
    file = open('timetable.txt', 'w')
    for key, value in classes.items():
        file.write(str(key) + str(value))
        file.write('\n')
    file.close()
        
elif save == '2':
    with open('timetable.pkl', 'wb') as file:
        pickle.dump(classes, file)

elif save == '3':
    file = open('timetable.txt', 'w')
    for key, value in classes.items():
        file.write(str(key) + str(value))
        file.write('\n')
    file.close()
    with open('timetable.pkl', 'wb') as file:
        pickle.dump(classes, file)
```
