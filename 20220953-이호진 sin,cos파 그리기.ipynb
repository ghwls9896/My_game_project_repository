```python
import math
import lee_ho_jin
def print_sin(start_deg,end_deg,length, width) :
    num_width = 0
    num_height = 0
    midd = width/2
    width += 1          
    print(str("x").rjust(2), end=" ")
    for num_width in range (width):
        print("-", end="")               #x---------------------sin(x)형태의 행을 미리 정의해줌
    print(str(" sin(x)").ljust(2))
    
    for num_height in range (start_deg,length):  #000행부터 시작
        x_val = num_height*10
        sin_val = (round(math.sin(math.pi * (x_val / (end_deg/2))), 3)) #맨 오른쪽 열에 나오는 sin(x)숫자 
        r_loc = round(midd + (midd * sin_val))          # *이 찍힐 위치를 sin(x)값을 이용하여 나타냄
        print(str(num_height).zfill(2), end="0")        #000, 010, ... 형태의 모양을 만들려면 문자열로 바꿔서 해주어야함
        print("", end="")
        for num_width in range (width):
            print(" ", end ="")
            if(sin_val == 0):
                if (num_width == midd):                #sin값이 0 일때 정 가운데 찍혀야할 *
                    print("*", end="")

            else:
                if(sin_val>=0):
                    if (num_width == midd):            
                        print("|", end="")
                else:                                       #폭의 절반인 가운데에 나와야할 |
                    if (num_width == midd-1):
                        print("|", end="")
                if(num_width == r_loc):                     #num_width가 위에서 정의한 r_loc 위치가 됐을때 실행
                    print("*", end="")
            if (num_width == width-1):                      
                print(str(format(sin_val, ".3f")), end = "")
                print("")
def print_cos(start_deg,end_deg,length, width) :
    num_width = 0
    num_height = 0
    midd = width/2
    width += 1          
    print(str("x").rjust(2), end=" ")
    for num_width in range (width):
        print("-", end="")                              #x---------------------sin(x)형태의 행을 미리 정의해줌
    print(str(" cos(x)").ljust(2))
    
    for num_height in range (start_deg,length):          #000행부터 시작
        x_val = num_height*10
        cos_val = (round(math.cos(math.pi * (x_val / (end_deg/2))), 3))           #맨 오른쪽 열에 나오는 cos(x)숫자 
        r_loc = round(midd + (midd * cos_val))                                       # *이 찍힐 위치를 cos(x)값을 이용하여 나타냄
        print(str(num_height).zfill(2), end="0")                         #000, 010, ... 형태의 모양을 만들려면 문자열로 바꿔서 해주어야함
        print("", end="")
        for num_width in range (width):
            print(" ", end ="")
            if(cos_val == 0):
                if (num_width == midd):                             #cos값이 0 일때 정 가운데 찍혀야할 *
                    print("*", end="")

            else:
                if(cos_val>=0):                                       #폭의 절반인 가운데에 나와야할 |
                    if (num_width == midd):
                        print("|", end="")
                else:
                    if (num_width == midd-1):
                        print("|", end="")
                if(num_width == r_loc):                                 #num_width가 위에서 정의한 r_loc 위치가 됐을때 실행
                    print("*", end="")
            if (num_width == width-1):
                print(str(format(cos_val, ".3f")), end = "")
                print("")
def main():
    print("헤더프린트에 들어갈 제목, 날짜, 이름을 정해주세요")
    title = input("제목을 적어주세요 :")
    date= input("날짜를 적어주세요 : ")
    name = input("이름을 적어주세요 : ")
    lee_ho_jin.header_print(title, date ,name)
    print("그래프의 시작 각도, 끝 각도, 그래프의 총 세로길이, 그래프의 가로 폭을 설정해주세요.")
    start_deg = int(input("그래프의 시작 각도 : (ex:0)"))
    end_deg = int(input("끝 각도 : (ex:360)"))
    length = int(input("그래프의 총 세로길이 : (ex:36)"))
    width = int(input("그래프의 가로 폭 : (ex:80)"))
    if(width>110 or width < 0):
        print("폭이 110이 넘거나 0보다 작으면 그래프가 정상적으로 나오지 않습니다.")
    else:
        print("sin함수 그래프")
        print_sin(start_deg, end_deg,length,width)
        for i in range(15):
            for j in range(3):
                print(".", end="")
            print("")
        print("cos함수 그래프")
        print_cos(start_deg,end_deg,length,width)
```
